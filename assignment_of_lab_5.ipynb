```
import math


x_A = [10]
y_A = [0]

x_B = [0]
y_B = [10]

x_C = [10]
y_C = [10]

x_D = [0]
y_D = [0]

v_A = 3
v_B = 5
v_C = 7
v_D = 2


D_gotShoots = 0
C_gotShoots = 0
B_gotShoots = 0


import math
def dis(x1, y1, x2, y2):
    return math.sqrt((x1 - x2) ** 2 + (y1 - y2) **2)




for i in range(20):

  distDA = dis(x_D[i], y_D[i], x_A[i], y_A[i])
  distCB = dis(x_C[i], y_C[i], x_B[i], y_B[i])
  distBD = dis(x_B[i], y_B[i], x_D[i], y_D[i])

  #print(distDA, distCB, distBD)

  print()
  print("At time : ", i)
  print("X,Y cordinate of car A :", x_A[i],",", y_A[i])
  print("X,Y cordinate of car B :", x_B[i],",", y_B[i])
  print("X,Y cordinate of car C :", x_C[i],",", y_C[i])
  print("X,Y cordinate of car D :", x_D[i],",", y_D[i])

  

  if distDA < 5:
    print('A shoots D, Distance DA : ', distDA)
    D_gotShoots +=1

  if distCB < 5:
    print('B shoots C, Distance CB :', distCB)
    C_gotShoots += 1

  if distBD < 5:
    print('D shots B, Distance BD: ', distBD)
    B_gotShoots +=1 





  y_C.append(y_C[i]+ v_C)
  x_C.append(x_C[0])

  #D is chassed by A
  sin = (y_D[i] - y_A[i]) / distDA
  cos =  ( x_D[i] - x_A[i] ) / distDA
  new_x_A =  x_A[i] + v_A * cos
  new_y_A =  y_A[i] + v_A * sin
  x_A.append(new_x_A)
  y_A.append(new_y_A) 



  #C is chased by B
  sin = (y_C[i] - y_B[i]) / distCB
  cos =  ( x_C[i] - x_B[i] ) / distCB
  new_x_B =  x_B[i] + v_B * cos
  new_y_B =  y_B[i] + v_B * sin
  x_B.append(new_x_B)
  y_B.append(new_y_B)


   #B is chased by D
  sin = (y_B[i] - y_D[i]) / distBD
  cos =  ( x_B[i] - x_D[i] ) / distBD
  new_x_D =  x_D[i] + v_D * cos
  new_y_D =  y_D[i] + v_D * sin
  x_D.append(new_x_D)
  y_D.append(new_y_D)

print()
print("D got shoot", D_gotShoots)
print("C got shoot", C_gotShoots)
print("B got shoot", B_gotShoots)








  

```
